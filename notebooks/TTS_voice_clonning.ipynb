```python
from TTS.api import TTS

# Running a multi-speaker and multi-lingual model

# List available 🐸TTS models and choose the first one
model_name = TTS()
model=model_name.list_models()[0]

# Init TTS
tts = TTS(model)
```

```python
tts.tts_to_file("There must be a vision leading to higher goal of achievement. you must dream, something i want to achieve in my life", speaker_wav="2_min_speech.wav", language="en", file_path="output_2_min.wav")
```

```python
import librosa
import noisereduce as nr
import soundfile as sf
# Load the audio file
audio_file = "output_kalam.wav"
y, sr = librosa.load(audio_file, sr=None)

# Perform noise reduction using noisereduce
reduced_noise = nr.reduce_noise(y=y, sr=sr)

# Save the cleaned audio
output_file = "cleaned_audio_kalam.wav"
sf.write(output_file, reduced_noise, sr)
```

```python
import librosa
import numpy as np

# Load the audio file
audio_file = "output_audio.wav"
y, sr = librosa.load(audio_file, sr=None)

# Compute speech rate (duration / number of frames)
duration = librosa.get_duration(y=y, sr=sr)
n_frames = len(y)
speech_rate = n_frames / duration
print("Speech Rate (frames/second):", speech_rate)

# Compute pitch (fundamental frequency)
# Compute pitch (fundamental frequency)
pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
mean_pitch = np.mean(pitches[np.argmax(magnitudes, axis=0)])
print("Mean Pitch (Fundamental Frequency):", mean_pitch)

# Compute spectral centroids
spectral_centroids = librosa.feature.spectral_centroid(y=y, sr=sr)
print("Spectral Centroids:", spectral_centroids)
```
